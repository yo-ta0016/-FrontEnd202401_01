generatetivemodeのモジュールのimportとAPIキーの設定

```python
import google.generativeai as genai
# ノートブックに設定したAPIキーをgenerativeaiに設定
from google.colab import userdata
genai.configure( api_key=userdata.get("GOOGLE_API_KEY"))
```

gemini hメッセージを送って応答取得

```python
model = genai.GenerativeModel("gemini-flash-latest")
response = model.generate_content("こんにちは！ 私は耀太と言います！")
print( response.text)
```

```python
print(response)
```

履歴を踏まえたメッセージの送受信

```python
from IPython.core import history
# モデルに履歴を設定して会話をスタートする
chat = model.start_chat(
    history=[
        {"role":"user","parts":["こんにちは！ 私は耀太と言います！"]},
        {"role":"model","parts":["こんにちは、耀太さん初めまして、何かお手伝いすることはありますか？"]}
    ]
)
# 会話にメッセージを追加して応答してもらう
response = chat.send_message("私の名前がわかりますか？")
print( response.text)
```

ストリーミング応答を得る

```python
# 上のセルのmodelを使う
response = model.generate_content("こんにちは！ 私は明太子耀太と言います。私がどこ出身かわかりますか？",
                                  stream = True)
# 応答を順番に表示する
for chunk in response:
  print(chunk.text, end="", flush=True)
  # forループ終わり
```

JSONモードでの応答

```python
import google.generativeai as genai
from google.colab import userdata

genai.configure(api_key=userdata.get("GOOGLE_API_KEY"))

model = genai.GenerativeModel('gemini-flash-latest')

response = model.generate_content(
    '昔々あるところにおじいさんとおばあさんがいました。登場人物を次のJSON形式で出力してください。{"people":["aaa","bbb"]}',
    generation_config={"response_mime_type":"application/json"}
)

print(response.text)
```

インターネットから画像を取得し、その内容を解析させる

```python
import google.generativeai as genai
import requests
from PIL import Image
from io import BytesIO

model = genai.GenerativeModel('gemini-flash-latest')

# Corrected direct image URL
image_url = "http://www.ponos.co.jp/iphone/img/battlecats/icon_nyanko.png"
response_img = requests.get(image_url)
img = Image.open(BytesIO(response_img.content))

response = model.generate_content(["この画像の内容を分析して.", img])
print(response.text)
```
